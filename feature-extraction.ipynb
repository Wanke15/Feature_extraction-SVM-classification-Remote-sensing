```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import caffe
```

```python
import sys
caffe_root = '../'  # this file should be run from {caffe_root}/examples (otherwise change this line)

import os
```

```python
caffe.set_mode_cpu()

model_def = caffe_root + 'models/bvlc_reference_caffenet/deploy.prototxt'
model_weights = caffe_root + 'models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel'

net = caffe.Net(model_def,      # defines the structure of the model
                model_weights,  # contains the trained weights
                caffe.TEST)     # use test mode (e.g., don't perform dropout)
```

```python
# load the mean ImageNet image (as distributed with Caffe) for subtraction
mu = np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy')
mu = mu.mean(1).mean(1)  # average over pixels to obtain the mean (BGR) pixel values
print 'mean-subtracted values:', zip('BGR', mu)

# create transformer for the input called 'data'
transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})

transformer.set_transpose('data', (2,0,1))  # move image channels to outermost dimension
transformer.set_mean('data', mu)            # subtract the dataset-mean value in each channel
transformer.set_raw_scale('data', 255)      # rescale from [0, 1] to [0, 255]
transformer.set_channel_swap('data', (2,1,0))  # swap channels from RGB to BGR
```

```python
# set the size of the input (we can skip this if we're happy
#  with the default; we can also change it later, e.g., for different batch sizes)
net.blobs['data'].reshape(1,        # batch size
                          3,         # 3-channel (BGR) images
                          227, 227)  # image size is 256x256
```

```python
net.blobs['data'].data.shape
```

```python
DATASET_DIR = 'data\\images\\UCMerced_LandUse\\Images\\'
```

```python
IMAGE_CLASSES = os.listdir(DATASET_DIR)
```

```python
IMAGE_CLASSES
```

```python
os.path.join(DATASET_DIR, IMAGE_CLASSES[0])
```

```python
remote_sensing_data = np.empty((2100, 4096))
```

```python
N = 0
for i, c in enumerate(IMAGE_CLASSES):
    #print i, c
    for img in os.listdir(os.path.join(DATASET_DIR+c)):
        image_dir = os.path.join(os.path.join(DATASET_DIR+c) ,img)
        print 'Processing class: {}'.format(image_dir)
        image = caffe.io.load_image(image_dir)
        #plt.imshow(image)
        transformed_image = transformer.preprocess('data', image)
        net.blobs['data'].data[...] = transformed_image
        net.forward()
        feat = net.blobs['fc7'].data[0].copy()
        remote_sensing_data[N] = feat
        N += 1
```

```python
plt.rcParams['figure.figsize'] = (10, 10)        # large images
plt.rcParams['image.interpolation'] = 'nearest'  # don't interpolate: show square pixels
plt.rcParams['image.cmap'] = 'gray'  # use grayscale output rather than a (potentially misleading) color heatmap

plt.plot(remote_sensing_data[1])
```
